```python
def validate_inputs(audience, tone):
    if audience not in valid_audiences:
        raise ValueError(f"Invalid audience: {audience}. Must be one of {valid_audiences}")

    if tone not in valid_tones:
        raise ValueError(f"Invalid tone: {tone}. Must be one of {valid_tones}")

    return audience, tone
```

# Task 6: Prompt Generator App

```python
teaching_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} teaching style"
)

interview_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Ask 3 interview questions about {topic} for {audience} in a {tone} tone"
)

story_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} storytelling style"
)
```

```python
valid_audiences = ["beginner", "intermediate", "expert"]
valid_tones = ["formal", "casual", "fun"]

def validate_inputs(audience, tone):
    if audience not in valid_audiences:
        audience = "beginner"
    if tone not in valid_tones:
        tone = "formal"
    return audience, tone
```

```python
def generate_prompt(topic, audience, tone, style):

    # Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Select template
    if style == "teaching":
        prompt = teaching_template
    elif style == "interview":
        prompt = interview_template
    elif style == "storytelling":
        prompt = story_template
    else:
        raise ValueError("Invalid style. Choose from: teaching, interview, storytelling")

    # Generate output
    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
print(generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
))
```

```python
print(generate_prompt("AI", "expert", "formal", "teaching"))
print(generate_prompt("Python", "intermediate", "casual", "interview"))
```

# Task 7: Template Reusability Test

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)
```

```python
test_cases = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "casual"),
    ("Deep Learning", "expert", "formal"),
    ("Data Science", "beginner", "formal"),
    ("Neural Networks", "intermediate", "fun")
]

for topic, audience, tone in test_cases:
    output = reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
    print(output)
```
